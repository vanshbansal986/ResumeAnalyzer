```python
import pandas as pd
import os
import json

def update_or_create_excel(file_path, new_data):
    """
    Updates or creates an Excel file with the given data.

    Args:
        file_path (str): Path to the Excel file.
        new_data (dict): Data to add to the Excel file.
    """
    # Convert the new data into a DataFrame
    new_df = pd.DataFrame([new_data])
    print("new_df: \n")
    print(new_df.info())
    
    # Check if the file exists
    if os.path.exists(file_path):
        # Read the existing Excel file
        try:
            existing_df = pd.read_excel(file_path)
            print("existing df: \n")
            print(new_df.info())
            # Append new data
            updated_df = pd.concat([existing_df, new_df], ignore_index=True)
        except Exception as e:
            print(f"Error reading the existing file: {e}")
            print("Creating a new file instead.")
            updated_df = new_df
    else:
        # If the file doesn't exist, create a new DataFrame
        updated_df = new_df

    # Save the updated DataFrame back to the Excel file
    print("updated df: \n")
    print(updated_df.info())
    #updated_df.to_excel(file_path, index=False)
    print(f"Data successfully saved to {file_path}")


# Example Usage
response = '''
{
    "name": "Vansh Bansal",
    "phone": null,
    "email": "vanshbansal986@gmail.com",
    "university": "Vellore Institute Of Technology",
    "year_of_study": "2025",
    "course": "Bachelor of Technology in Information Technology",
    "cgpa": "8.4",
    "Key Skills": ["C", "C++", "Python", "HTML", "SQL", "NumPy", "Pandas", "Matplotlib", "BeautifulSoup", "Selenium", "Tableau", "Sklearn", "Git", "LangChain", "Data Structures and Algorithms", "Data Analysis", "Machine Learning", "GenAI", "LLMs", "OpenAI", "HuggingFace"],
    "Gen AI Experience Score": 2,
    "AI/ML Experience Score": 3,
    "supporting_information": "Implemented data scraping and analysis in internship at CollabJ to identify leads. Designed end-to-end ML project for network security data using MongoDB, MLFlow, and Docker. Developed influencer insights system with face analysis. Conducted credit score modeling with Sklearn and feature selection techniques. Built energy price prediction model with PCA and GridSearchCV."
}
'''

# Convert the JSON string to a Python dictionary
new_data = json.loads(response)

# Specify the Excel file path
file_path = "data.xlsx"

# Call the function to update or create the Excel file
update_or_create_excel(file_path, new_data)
```
